```python
from sklearn.datasets import make_classification
```

```python
X,y=make_classification(n_samples=1000,n_redundant=0,n_features=2,n_clusters_per_class=1,weights=[.90],random_state=12)
```

```python
import pandas as pd
df1=pd.DataFrame(X,columns=['f1','f2'])
df2=pd.DataFrame(y,columns=['target'])
df_total=pd.concat([df1,df2],axis=1)
df_total.head()
```

```python
df_total['target'].value_counts()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df_total['f1'],df_total['f2'],c=df_total['target'])
```
