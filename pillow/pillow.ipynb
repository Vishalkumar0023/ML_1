```python
from PIL import Image

# open an image file
img = Image.open("py.jpeg")

# show image
img.show()
```

```python
print("Format:", img.format)       # e.g. JPEG, PNG
print("Mode:", img.mode)           # e.g. RGB, L (grayscale)
print("Size:", img.size)           # (width, height)
```

```python
small = img.resize((400, 200))
small.show()
```

```python
rotated = img.rotate(45)   # rotate 45 degrees
rotated.show()
```

```python
# Crop region (left, top, right, bottom)
cropped = img.crop((100, 100, 300, 300))
cropped.show()
```

```python
# flipped = img.transpose(Image.FLIP_LEFT_RIGHT)
# flipped.show()

flop = img.transpose(Image.FLIP_TOP_BOTTOM)
flop.show()
```

```python
gray = img.convert("L")
gray.show()
```

```python
gray.save("gray_sample.jpg")
```

```python
from PIL import Image, ImageFilter

# Open image
img = Image.open("py.jpeg")

# Convert to RGB if it’s in palette mode
if img.mode == "P":
    img = img.convert("RGB")

# Apply blur filter
blurred = img.filter(ImageFilter.BLUR)

# Show blurred image
blurred.show()
```
