```python
import gc

class MyObject:
    def __init__(self,name):
        self.name=name
        print(f"Object {self.name} is created")
    def __del__(self):
        print(f"Object {self.name} is deleted")
obj1=MyObject("obj1")
obj2=MyObject("obj2")

# circular refernce
obj1.ref = obj2
obj2.ref = obj1

# manaual trigger garbage collection
gc.collect()
        
```

```python
## generators

def generate_number(n):
    for i in range(n):
        yield i

for num in generate_number(1000):
    print(num)
    if num>10:
        break        
```

```python
import tracemalloc

def create_list():
    return [i for i in range(10000)]
def main():
    tracemalloc.start()

    create_list()
    snapshot=tracemalloc.take_snapshot()
    top_stats =snapshot.statistics('lineno')
    print("[top 10]")
    for stat in top_stats[:10]:
        print(stat)
```

```python
if __name__ == "__main__":
    main()
```
